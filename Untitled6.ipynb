```python
# Anomally detection

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
df = pd.read_csv("/content/50k.csv")

# Convert 'time' to datetime and remove duplicates
df["time"] = pd.to_datetime(df["time"], errors="coerce")
df.drop_duplicates(inplace=True)

# Handle missing values
df.fillna(df.select_dtypes(include=[np.number]).median(), inplace=True)

# Feature selection (Remove highly correlated features)
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.85) and column != "t2m"]
df.drop(columns=to_drop, inplace=True)

# 🚀 Overfitting Reduction (Apply Gaussian Noise to 't2m')
df["t2m"] += np.random.normal(loc=0, scale=0.3, size=len(df))

# Generate 'anomaly' column using a more relaxed threshold
threshold_high = df["t2m"].quantile(0.968)  # Adjusted from 0.975
threshold_low = df["t2m"].quantile(0.032)   # Adjusted from 0.025
df["anomaly"] = ((df["t2m"] > threshold_high) | (df["t2m"] < threshold_low)).astype(int)

# Check class distribution
anomaly_count = df["anomaly"].sum()
normal_count = len(df) - anomaly_count
print(f"Anomalies: {anomaly_count}, Normal: {normal_count}")

# Balance dataset
sample_size = min(anomaly_count, normal_count)
if sample_size == 0:
    raise ValueError("No anomalies found! Check your thresholding method.")

df_normal = df[df["anomaly"] == 0].sample(sample_size, random_state=42)
df_anomaly = df[df["anomaly"] == 1].sample(sample_size, random_state=42)
df_balanced = pd.concat([df_normal, df_anomaly]).sample(frac=1, random_state=42)

# Define features & target
features = ["latitude", "longitude", "t2m", "u100", "v100"]
target = "anomaly"

X = df_balanced[features]
y = df_balanced[target]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=42)

# 🚀 Improved Model (Better Generalization)
model = RandomForestClassifier(n_estimators=10, max_depth=2, min_samples_split=10, min_samples_leaf=12,
                               class_weight="balanced", random_state=42)
model.fit(X_train, y_train)

# Predictions
y_prob = model.predict_proba(X_test)[:, 1]  # Get probability scores

# 🔹 Find Best Threshold
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]

# Apply New Threshold
y_pred = (y_prob >= optimal_threshold).astype(int)

# Performance Metrics
train_acc = accuracy_score(y_train, model.predict(X_train))
test_acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=1)
recall = recall_score(y_test, y_pred, zero_division=1)
f1 = f1_score(y_test, y_pred, zero_division=1)

# Print results
print("\n===== Training & Testing Scores =====")
print(f"Training Accuracy: {train_acc:.4f}")
print(f"Testing Accuracy: {test_acc:.4f}")
print(f"Total Model Accuracy: {(train_acc + test_acc) / 2:.4f}")

print("\n===== Performance Metrics =====")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# 🚀 First 10 Anomalies
anomalies = df_balanced[df_balanced["anomaly"] == 1][["time", "latitude", "longitude", "t2m"]].head(10)
print("\n===== First 10 Anomalies Detected =====")
print(anomalies)

# =========================== VISUALIZATIONS ===========================

# 2️⃣ Scatter Plot - Latitude vs Temperature with Anomalies
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_balanced,
    x="latitude",
    y="t2m",
    hue="anomaly",
    palette={0: "blue", 1: "red"},
    alpha=0.7
)
plt.title("Scatter Plot - Latitude vs Temperature with Anomalies")
plt.xlabel("Latitude")
plt.ylabel("Temperature (t2m)")

# Fix legend colors
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Normal', markersize=8, markerfacecolor='blue'),
                   Line2D([0], [0], marker='o', color='w', label='Anomaly', markersize=8, markerfacecolor='red')]
plt.legend(handles=legend_elements, title="Anomaly", loc="best")

plt.grid(True)
plt.show()

# 1️⃣ Feature Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_balanced[features].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Climate Change Prediction

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, LayerNormalization, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load dataset
df = pd.read_csv("/content/50k.csv")  # Replace with actual dataset path
df.columns = df.columns.str.strip()

# Select relevant features
selected_features = ["latitude", "longitude", "t2m", "u100", "v100", "msl"]
df = df[selected_features].dropna()

# Normalize data
t2m_scaler = MinMaxScaler()
df["t2m"] = t2m_scaler.fit_transform(df[["t2m"]])

feature_scaler = MinMaxScaler()
df_scaled = feature_scaler.fit_transform(df.drop(columns=["t2m"]))

# Prepare sequences
time_steps = 48
X, y = [], []
for i in range(len(df_scaled) - time_steps):
    X.append(df_scaled[i:i + time_steps])
    y.append(df.iloc[i + time_steps]["t2m"])  # Keep t2m unscaled

X, y = np.array(X), np.array(y)

# Train-test split
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build LSTM model
model = Sequential([
    Bidirectional(LSTM(128, return_sequences=True), input_shape=(time_steps, X.shape[2])),
    LayerNormalization(),
    Dropout(0.25),

    Bidirectional(LSTM(64, return_sequences=False)),
    LayerNormalization(),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')
])

# Compile model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='mean_squared_error')

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
]

# Train model
history = model.fit(X_train, y_train, epochs=100, batch_size=256, validation_data=(X_test, y_test), callbacks=callbacks)

# Predict with LSTM
lstm_predictions = model.predict(X_test)

# Train Random Forest
rf_model = RandomForestRegressor(n_estimators=300, max_depth=15)
rf_model.fit(lstm_predictions, y_test[:len(lstm_predictions)])

# Predict with Random Forest
final_predictions = rf_model.predict(lstm_predictions)

# **Inverse transform t2m values to Kelvin**
final_predictions_original = t2m_scaler.inverse_transform(final_predictions.reshape(-1, 1)).flatten()

# **Extract Latitude and Longitude from dataset**
timestamps = pd.date_range(start="2025-01-01", periods=10, freq="D")
first_10_results = pd.DataFrame({
    "Time": timestamps.strftime("%Y-%m-%d"),
    "Latitude": df.iloc[split:split+10]["latitude"].values,
    "Longitude": df.iloc[split:split+10]["longitude"].values,
    "t2m": np.round(final_predictions_original[:10], 5)  # Rounded to 5 decimal places
})

# **Print results in exact required format**
print("\nFirst 10 Predicted Climate Values for 2025 (Actual Scale in Kelvin):")
print(first_10_results)

# =========================== Model Performance Metrics ===========================

rmse = np.sqrt(mean_squared_error(y_test[:len(lstm_predictions)], final_predictions))
mae = mean_absolute_error(y_test[:len(lstm_predictions)], final_predictions)

# Display results
print("\n===== Training & Testing Loss =====")
print(f"LSTM Training Loss (MSE): {history.history['loss'][-1]:.4f}")
print(f"LSTM Testing Loss (MSE): {history.history['val_loss'][-1]:.4f}")
print(f"Random Forest Training Loss (MSE): {mean_squared_error(y_train[:len(lstm_predictions)], rf_model.predict(model.predict(X_train[:len(lstm_predictions)]))):.4f}")
print(f"Random Forest Testing Loss (MSE): {mean_squared_error(y_test[:len(lstm_predictions)], final_predictions):.4f}")

print("\n===== Final Performance Metrics =====")
print(f"RMSE: {rmse:.5f}")
print(f"MAE: {mae:.5f}")

# =========================== Bar Chart - Model Performance ===========================

metrics = {
    "LSTM Train Loss": history.history['loss'][-1],
    "LSTM Test Loss": history.history['val_loss'][-1],
    "RF Train Loss": mean_squared_error(y_train[:len(lstm_predictions)], rf_model.predict(model.predict(X_train[:len(lstm_predictions)]))),
    "RF Test Loss": mean_squared_error(y_test[:len(lstm_predictions)], final_predictions),
    "RMSE": rmse,
    "MAE": mae
}

plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'blue', 'green', 'green', 'red', 'red'], alpha=0.7)
plt.xlabel("Metrics")
plt.ylabel("Error Value")
plt.title("Model Performance Comparison")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()

# =========================== VISUALIZATIONS ===========================

# 1️⃣ Annotated Time-Series Plot (Predicted vs. Actual)
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), t2m_scaler.inverse_transform(y_test.reshape(-1, 1)), label="Actual Temperature", color="blue", alpha=0.6)
plt.plot(range(len(final_predictions_original)), final_predictions_original, label="Predicted Temperature", color="red", linestyle="dashed")
plt.xlabel("Time Step")
plt.ylabel("Temperature")
plt.legend()
plt.title("Annotated Time-Series Plot - Climate Predictions")
plt.show()

# 2️⃣ Heatmap - Feature Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
